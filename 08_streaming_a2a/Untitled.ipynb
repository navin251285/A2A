```python
import asyncio
import websockets
import json

async def run():
    uri = "ws://localhost:9000/ws"   # ⚠️ also fix your port (you used 9001)

    async with websockets.connect(
        uri,
        origin="http://localhost"   # ✅ FIX
    ) as websocket:

        await websocket.send(json.dumps({
            "user_input": "calculate interest for 1000 at 5 for 2 years"
        }))

        while True:
            try:
                msg = await websocket.recv()
                print("📡", json.loads(msg))
            except:
                break


```

```python
await run()
```
