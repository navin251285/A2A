#### Lets do 1. Streaming HTTP Response

#####  Non streaming response means Client → Request → Server → (wait...) → Full Response

##### Streaming response means Client → Request → Server → (send data in chunks continuously)

```python
import requests

url = "http://localhost:9000/stream"

params = {
    "user_input": "calculate interest for 1000 at 5 for 2 years"
}

response = requests.get(url, params=params, stream=True)

for chunk in response.iter_lines():
    if chunk:
        print(chunk.decode())
```

```python
import requests
import json

url = "http://localhost:9001/stream"

params = {
    "user_input": "calculate interest for 1000 at 5 for 2 years"
}

response = requests.get(url, params=params, stream=True)

for line in response.iter_lines():
    if line:
        decoded = line.decode()

        if decoded.startswith("data: "):
            data = json.loads(decoded.replace("data: ", ""))
            print("📡 EVENT:", data)
```
