```python
import requests

res = requests.post(
    "http://localhost:9000/execute",
    params={"query": "add 10 and 20"}
)

```

```python
print("STATUS:", res.status_code)
print("RAW RESPONSE:", res.text)
print(res.json())
```

```python
requests.get("http://localhost:9000/result").json()
```
