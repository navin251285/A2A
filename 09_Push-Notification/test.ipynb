```python
import requests
import time

BASE = "http://localhost:9000"

# STEP 1: Submit
res = requests.post(
    f"{BASE}/execute",
    params={"query": "add 10 and 20"}
)

data = res.json()
print("🚀 SUBMIT:", data)

task_id = data.get("task_id")

if not task_id:
    print("❌ ERROR:", data)
    exit()

# STEP 2: Optional polling (fallback)
while True:
    r = requests.get(f"{BASE}/result/{task_id}")
    result = r.json()

    print("📊 STATUS:", result)

    if result.get("status") == "completed":
        print("🎉 FINAL RESULT:", result["result"])
        break

    time.sleep(1)
```
